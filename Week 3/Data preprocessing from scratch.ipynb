# checking datatype

```python
# Open the file for reading
file = open('data.csv', 'r')

# Read the first line of the file to get the column names
header = file.readline().strip().split(',')

# Create empty lists for each column
data = {}
for col in header:
    data[col] = []

# Loop over the remaining lines in the file
for line in file:
    # Split the line into a list of values using a comma as the delimiter
    values = line.strip().split(',')
    
    # Loop over the columns and append the value to the appropriate list
    for i, col in enumerate(header):
        if values[i] == '':
            data[col].append(None)
        elif values[i].isnumeric():
            data[col].append(int(values[i]))
        elif '.' in values[i]:
            data[col].append(float(values[i]))
        else:
            data[col].append(values[i])
        
# Print the datatype of each column
for col in header:
    unique_types = set([type(value).__name__ for value in data[col]])
    if len(unique_types) == 1:
        print('Column {} has datatype {}.'.format(col, list(unique_types)[0]))
    else:
        print('Column {} has mixed datatypes.'.format(col))
 
```

# checking null values

```python
with open('data.csv', 'r') as file:

    # Read the first line of the file
    header = file.readline().rstrip().split(',')

    # Initialize a dictionary to count null values for each column
    null_counts = {col: 0 for col in header}

    # Iterate over each subsequent line in the file
    for line in file:
        # Split the line into its values
        values = line.rstrip().split(',')

        # Increment the null count for each null value in the line
        for col_num, value in enumerate(values):
            if not value:
                null_counts[header[col_num]] += 1

    # Print the null counts for each column
    for col, count in null_counts.items():
        print(f'{col}: {count}')
```

# Imputing missing values

```python
# Open input file and read data
with open('data.csv', 'r') as file:
    data = [line.strip().split(',') for line in file]

# Find mean of column
col_num = 3  # Index of column to impute
col_data = [int(row[col_num]) for row in data[1:] if row[col_num]]
mean_val = round(sum(col_data) / len(col_data))

# Impute mean value
for row in data[1:]:
    if not row[col_num]:
        row[col_num] = str(mean_val)

# Write updated data to new file
with open('imputed_data.csv', 'w') as file:
    for row in data:
        file.write(','.join(row) + '\n')
```

# converting categorical to numerical

```python
category_map = {'male': 1, 'female': 0}

with open('imputed_data.csv', 'r') as file1:
    data1 = file1.readlines()
#Open new file for writing
with open('updated_data.csv', 'w') as file2:
    # Loop over data1 and update data
    for line in data1:
        values = line.strip().split(',')
        for i in range(len(values)):
            if values[i] in category_map:
                values[i] = str(category_map[values[i]])
        
        # Write updated row to file2
        updated_row = ','.join(values)
        file2.write(updated_row + '\n')
```
